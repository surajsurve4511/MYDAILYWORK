```python
from datetime import datetime
import random

print("Chatbot: Hello! Iâ€™m RuleBot ðŸ¤–. Whatâ€™s your name?")

# State
user_name = None
last_topic = None
mood = "neutral"

while True:
    user_input = input("You: ").lower()

    # Exit
    if user_input in ["bye", "exit", "quit", "goodbye"]:
        print("Chatbot: Goodbye! Take care! ðŸ‘‹")
        break

    # Name detection
    elif user_name is None and "my name is" in user_input:
        user_name = user_input.split("my name is")[-1].strip().capitalize()
        print(f"Chatbot: Nice to meet you, {user_name}!")
    elif user_name is None and "i am" in user_input:
        user_name = user_input.split("i am")[-1].strip().capitalize()
        print(f"Chatbot: Got it, {user_name}!")
    elif user_name is None and "i'm" in user_input:
        user_name = user_input.split("i'm")[-1].strip().capitalize()
        print(f"Chatbot: Hello {user_name}!")

    elif user_name is None:
        print("Chatbot: Please tell me your name first. Try saying 'My name is John'.")

    # Greetings
    elif "hello" in user_input or "hi" in user_input or "hey" in user_input:
        print(f"Chatbot: Hello {user_name}! How can I help you today?")

    # Time and Date
    elif "time" in user_input:
        print(f"Chatbot: Itâ€™s {datetime.now().strftime('%H:%M:%S')}.")
        last_topic = "time"
    elif "date" in user_input or "day" in user_input:
        print(f"Chatbot: Today is {datetime.today().strftime('%A, %Y-%m-%d')}.")
        last_topic = "date"

    # Weather (simulated)
    elif "weather" in user_input or "rain" in user_input or "sunny" in user_input or "temperature" in user_input:
        print("Chatbot: I can't access live weather, but I hope it's nice where you are!")
        last_topic = "weather"

    # Joke
    elif "joke" in user_input or "make me laugh" in user_input or "funny" in user_input:
        jokes = [
            "Why do programmers prefer dark mode? Because the light attracts bugs!",
            "Parallel lines have so much in common... itâ€™s a shame theyâ€™ll never meet.",
            "I asked the computer for a joke, but it just gave me a 404."
        ]
        print("Chatbot:", random.choice(jokes))
        last_topic = "joke"

    # Gratitude
    elif "thank" in user_input or "thanks" in user_input:
        print(f"Chatbot: You're very welcome, {user_name}!")

    # Emotion
    elif "i feel" in user_input or "i am feeling" in user_input or "i'm feeling" in user_input:
        if "sad" in user_input:
            mood = "sad"
            print("Chatbot: I'm here for you. Want a joke or a motivational quote?")
        elif "happy" in user_input:
            mood = "happy"
            print("Chatbot: Thatâ€™s great! Keep smiling ðŸ˜„")
        elif "angry" in user_input:
            mood = "angry"
            print("Chatbot: Try taking a deep breath. Iâ€™m here to listen.")
        else:
            print("Chatbot: I'm glad you shared that.")

    # Motivation
    elif "motivate" in user_input or "motivation" in user_input or "inspire" in user_input:
        quotes = [
            "Believe you can and you're halfway there.",
            "The best time to start was yesterday. The second best time is now.",
            "Youâ€™re capable of amazing things!"
        ]
        print("Chatbot:", random.choice(quotes))

    # Favorite color/food
    elif "favorite color" in user_input:
        print("Chatbot: I love all colors equally, but I'm told blue suits me best. ðŸ’™")
    elif "favorite food" in user_input:
        print("Chatbot: I don't eat, but I think pizza sounds amazing!")

    # Knowledge / fun facts
    elif "fact" in user_input or "tell me something" in user_input:
        facts = [
            "Did you know? Honey never spoils.",
            "Octopuses have three hearts.",
            "Bananas are berries, but strawberries aren't!"
        ]
        print("Chatbot:", random.choice(facts))

    # Do you love me
    elif "do you love me" in user_input or "are you my friend" in user_input:
        print("Chatbot: Of course! Iâ€™m your loyal chatbot friend ðŸ’›")

    # How are you?
    elif "how are you" in user_input:
        print("Chatbot: Iâ€™m doing well, especially when chatting with you!")

    # What can you do?
    elif "what can you do" in user_input or "help" in user_input or "options" in user_input:
        print("Chatbot: I can tell jokes, give the time/date, talk about moods, share facts, and more!")

    # Reminders (simulated)
    elif "remind me" in user_input or "reminder" in user_input:
        print("Chatbot: I wish I could set real reminders, but Iâ€™m not connected to your system.")

    # Age
    elif "how old are you" in user_input:
        print("Chatbot: I was created recently, but Iâ€™m wise beyond my lines of code!")

    # Are you real?
    elif "are you real" in user_input:
        print("Chatbot: I'm as real as a line of code can be. Not human, but 100% here for you!")

    # Bored
    elif "i am bored" in user_input or "i'm bored" in user_input:
        print("Chatbot: Want a fun fact or a joke to lift your boredom?")
        last_topic = "boredom"

    # Love
    elif "love" in user_input:
        print("Chatbot: Love is powerful. I may not feel it, but I can talk about it!")

    # Fallback
    else:
        if last_topic == "joke":
            print("Chatbot: Want another joke? Just say 'joke' again!")
        elif last_topic == "weather":
            print("Chatbot: You can ask 'Is it sunny?' or 'Is it raining?'")
        elif last_topic == "boredom":
            print("Chatbot: Type 'fact' or 'joke' to cure your boredom!")
        else:
            print("Chatbot: I didnâ€™t quite catch that. Try asking something else or type 'help'.")
```
